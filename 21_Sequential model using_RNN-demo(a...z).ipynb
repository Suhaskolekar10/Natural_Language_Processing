```python
import numpy as np
```

```python
alphabets = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
```

```python
len(alphabets)
```

### Process the data

```python
for i,v in enumerate(alphabets):
    print(i,v)
```

```python
int_to_char = dict(enumerate(alphabets))

int_to_char
```

```python
char_to_int = dict((v,i) for i,v in enumerate(alphabets))

char_to_int
```

```python
int_to_char[5]
```

```python
char_to_int['P']
```

```python
#Sequence modelling

seq_length = 1

x = []
y = []

for i in range(len(alphabets)-seq_length):
    seq_in = [alphabets[i]]
    seq_out = alphabets[i+1]
    x += [[char_to_int[seq_in[0]]]]
    y += [char_to_int[seq_out]]
    print(seq_in,"--->",seq_out)
```

### One hot encoding

```python
from keras.utils import to_categorical  #to match the no of classes
```

```python
y_new =to_categorical(y)
```

```python
y_new
```

```python
y_new.shape
```

### Scaling the input data

```python
np.array(x).shape
```

```python
x = np.reshape(x,(25,1,1))
```

```python
x.shape
```

```python
x
```

# when min number is zero to normalize we can divide it by maximum number

```python
x = x/25    #here we take 25 because 1 should be in output values
```

```python
x
```

### Build the model

```python
from keras.models import Sequential
from keras.layers import Dense,LSTM
```

```python
model = Sequential()

model.add(Dense(25, input_shape =(1,1), activation = 'relu'))
model.add(LSTM(128))
model.add(Dense(26,activation = 'softmax'))        #here we take 26 because for A it will generare automatically
```

```python
model.compile(loss = 'categorical_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
```

```python
model.fit(x,y_new,epochs=2000,batch_size=1)
```

```python
model.evaluate(x,y_new,batch_size=1)
```

### predict on the new data

```python
new = 'Z'
```

```python
new = char_to_int[new]
new = np.reshape(new,(1,1,1))
new = new/25
result = np.argmax(model.predict(new,verbose=False))
print(int_to_char[result])
```
