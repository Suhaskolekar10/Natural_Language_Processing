### Stopwords:-

1. Stop words are typically function words, which are words with little or no meaning that help form a sentence. Examples of stop words in English include "a", "the", "is", "are", "in", "at", "which", "on", etc.
2. Stop words are commonly used words that are filtered out before or after processing of natural language data (text) because they are considered insignificant. 
3. The reason why stop words are critical to many applications is that by removing the words that are very commonly used in a given language, we can focus on the important words instead. 
    
    For example, in the context of a search engine, if your search query is "how to develop information retrieval applications", the search engine can disregard the terms "how", "to" because these terms are so commonly used in the English language. This allows the search engine to focus on retrieving pages that contain the keywords: "develop", "information", "retrieval", "applications" â€“ which would bring up pages that are actually of interest 4.

### Punctuation removal:-

1. ***Clarity and Understanding***: Punctuation marks provide structure to sentences and highlight the tone of the message. They help the reader understand the context and meaning of the text.
2. ***Reduction of Noise***: Punctuation marks are not always meaningful in the context of text analysis. For example, in a text classification task, punctuation marks may not provide any additional information about the class of the text. Therefore, removing punctuation marks can help to reduce the noise and make the data cleaner and easier to work with.
3. ***Consistency***: Text data often comes from different sources, and these sources may use different styles of punctuation. Removing punctuation marks can help to standardize the data and make it more consistent.
4. ***Preservation of Meaning***: Some punctuation marks can change the meaning of a sentence. Removing punctuation marks can help to preserve the original meaning of the text.

*In conclusion, while punctuation marks are important for understanding and conveying meaning, they can be removed in NLP tasks to reduce noise, standardize the data, and preserve the original meaning. However, the decision to remove punctuation marks should be based on the specific requirements of the task.*

```python
text = '''Virat Kohli, born 5 November 1988 is an Indian international cricketer and 
the former captain of the Indian national cricket team. 
He currently represents Royal Challengers Bangalore in the IPL and Delhi in domestic cricket. 
Kohli is widely regarded as one of the greatest batsmen in the history of the sport, and the best of this era.
He is the highest run scorer in T20I and IPL. In 2020, the International Cricket Council named him the male 
cricketer of the decade. Kohli is currently fourth-highest run-scorer in international cricket and stands 
second in the list of most international centuries scored. 
He also holds the record for scoring the most centuries in One Day International cricket.
Kohli was a member of the Indian team that won the 2011 Cricket World Cup and 2013 ICC Champions Trophy.'''
```

```python
# imports

from nltk.tokenize import word_tokenize
import string
from nltk.corpus import stopwords
```

```python
# Remove punctuation marks from the text
punc = string.punctuation  #storing punctuations in the punc variable

text_token =  word_tokenize(text)     #Tokenizing the text

punc_free_text = []
for word in text_token:
    if (word not in punc and word.isalnum()):
        punc_free_text.append(word)

punc_free_text
```

```python
#Removing the stopwords

pure_text = []
for word in punc_free_text:
    if word.lower() not in stopwords.words('english'):
        pure_text.append(word)

pure_text
```

```python
#now plot the frequency plot   (it will show how frequently any word is present in the text)

from nltk import FreqDist

freq = FreqDist(word_tokenize(text))       #on plain text
freq.plot(20,cumulative=False);
```

```python
freq = FreqDist(punc_free_text)       #on punc free text
freq.plot(20,cumulative=False);
```

```python
freq = FreqDist(pure_text)       #on punc free text
freq.plot(20,cumulative=False);
```

```python
#All above steps in single function using list comphrension

def freq_plot(text):                      #function is for plain text
    token1 = word_tokenize(text)
    token2 = [word for word in token1 if word not in string.punctuation]
    token3 = [word for word in token2 if word not in stopwords.words('english')]
    
    freq = FreqDist(token1)
    freq.plot(10,cumulative=False)
    
    freq = FreqDist(token2)
    freq.plot(10,cumulative=False)
    
    freq = FreqDist(token3)
    freq.plot(10,cumulative=False)
```

```python
freq_plot(text)
```
