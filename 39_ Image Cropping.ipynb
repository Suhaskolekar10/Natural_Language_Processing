```python
img = cv2.imread('caves.jpg')
rgb_img =  cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(rgb_img)
```

```python
img.size
```

```python
#### Relative Cropping
```

```python
height, width, _ = img.shape
```

```python
0.15*height
```

```python
0.15*width
```

```python
#15% cropping from both the sides
```

```python
new_img=img[round(0.15*height):round(0.85*height),round(0.15*width):round(0.85*width)]     #slicing method for cropping an image
```

```python
plt.imshow(new_img)
```

```python
new_img.shape
```

```python
plt.figure(figsize=(16,9))
plt.subplot(1,2,1)
plt.grid()
plt.title('Original Image:' + str(rgb_img.shape))
plt.imshow(rgb_img)
plt.subplot(1,2,2)
plt.title('Cropped Image'+ str(new_img.shape))
plt.imshow(new_img)
```
