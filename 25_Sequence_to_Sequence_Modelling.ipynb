```python
headlines = []
for filename in os.listdir('/content/dive/MyDrive/Articles'):
  if "Articles" in filename:
    article_df = pd.read_csv('/content/dive/MyDrive/Articles/'+filename)
    headlines.extend(list(article_df['headline'].values))
```

```python
headlines = [h for h in headlines if h != 'Unknown']
print("the number of headliness is: ",len(headlines))
```

```python
import string
import warnings
warnings.filterwarnings('ignore')
warnings.simplefilter(action ='ignore', category = FutureWarning)
```

```python
def clean_text(headline):
  text = "".join( word for word in headline if word not in string.punctuation).lower()
  text = text.encode('utf8').decode("ascii","ignore")
  return text

corpus = [clean_text(headline) for headline in headlines]
```

```python
corpus
```

```python
vocab = []
for lines in corpus:
  words =lines.split()
  for word in words:
    vocab.append(word)

vocabulary = set(vocab)
```

```python
len(vocabulary)
```

```python
from keras.preprocessing.text import Tokenizer
from keras_preprocessing.sequence import pad_sequences
from keras.utils import to_categorical


tokenizer = Tokenizer()
tokenizer.fit_on_texts(corpus)
word2index = tokenizer.word_index
len(word2index)
```

```python
word2index
```

```python
dictionary = {}
rev_dictionary = {}
for word, idx in word2index.items():
  dictionary[word] =idx
  rev_dictionary[idx] =word
```

```python
rev_dictionary
```

```python
input_sequences = tokenizer.texts_to_sequences(corpus)

input_sequences
```

```python
input_data = []
target = []
for line in input_sequences:
  for i in range(1,len(line)-1):
    input_data.append(line[:i])
    target.append(line[i])
```

```python
#for padding we have to find maximum length

MAX_LEN = 0
for seq in input_data:
  if len(seq)>MAX_LEN:
    MAX_LEN = len(seq)

MAX_LEN
```

```python
input_data = pad_sequences(input_data,maxlen = MAX_LEN, padding="post")
```

```python
input_data.shape
```

```python
target = to_categorical(target, num_classes=len(word2index))
```

```python
target.shape
```

```python
#Build the model
from keras.models import Sequential
from keras.layers import Embedding, GRU, Dense, Dropout, LSTM
from keras.callbacks import EarlyStopping

```

```python
model = Sequential()

model.add(Embedding(input_dim=len(vocabulary), output_dim= 100, input_length = MAX_LEN))

model.add(LSTM(units = 100))

model.add(Dropout(rate = 0.1))

model.add(Dense(units=target.shape[1], activation = 'softmax'))
```

```python
model.compile(loss='categorical_crossentropy',optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(input_data, target, batch_size=10, epochs=20)
```

```python
new_sent = 'we need'

new_sent1 = 'Trump'
```

```python
def predict(text):
  input_sequences = tokenizer.texts_to_sequences([text])
  input_data = pad_sequences(input_sequences,maxlen = MAX_LEN, padding="post")

  result = np.argmax(model.predict(input_data, verbose=False))

  return rev_dictionary[result]
```
